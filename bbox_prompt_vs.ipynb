```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path = "/media/Datacenter_storage/PublicDatasets/cerebral_microbleeds_MAYO/mayo_t2s_png/images/test/10734610_11.png"
image = Image.open(img_path)
image_array = np.array(image)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,10))

# Display the image
ax.imshow(image_array, cmap='gray')

x_min = 311/4
x_max = 732/4
y_min = 542/4
y_max = 844/4

# Calculate width and height
width = x_max - x_min
height = y_max - y_min

# Create and add the rectangle
rect = plt.Rectangle((x_min, y_min), width, height, linewidth=5, edgecolor='red', facecolor='none')
ax.add_patch(rect)

# Add labels and title
ax.set_title(f'Image with Bounding Box\nBbox: [{x_min}, {y_min}, {x_max}, {y_max}]', fontsize=14)
ax.set_xlabel('X (pixels)')
ax.set_ylabel('Y (pixels)')

plt.tight_layout()
plt.show()
```
